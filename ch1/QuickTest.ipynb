```python
import matplotlib.pyplot as plt
%matplotlib inline



def wrap():
    
    for alpha in [-0.5, 1.5, 2.0, 2.5]:
        q0 = 0
        out = [q0]
        for r in [1, 1, 1, 1, 1, 1]:
            q0 = rec(q0, r, alpha)
            out.append(q0)
        plt.grid()
        plt.plot(out)
        plt.show()

        
def rec(q, r, alpha):
    return (q + alpha * (r-q))
```

```python
wrap()
```
